# Data Visualisation

```python
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4,5])
y = x**2+3
```

```python
plt.style.available
```

```python
plt.style.use('seaborn')
```

```python
print(x,y)
```

```python
plt.plot(x,y,label="Stock price 1")
plt.plot(x,x**2,label="Stock price 2")
plt.title("My line Plot")
plt.xlabel("TimeStamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,label="Stock price 1")
plt.scatter(x,x**2,label="Stock price 2")
plt.title("My line Plot")
plt.xlabel("TimeStamp")
plt.ylabel("Price")
plt.legend()
plt.show()
```

# Open CV

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Downloads/lamborghini_essenza_scv12_2020_4k_5k_hd_2-5120x2880.jpg')
```

```python
print(img.shape)
```

```python

plt.imshow(img)
```

```python
rgb_image = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
crop_image = rgb_image[700:2000,1000:4000,:]
plt.imshow(crop_image)
plt.show()
```

```python
crop_image.shape
```

```python
total_pixels = crop_image.shape[0]*crop_image.shape[1]
```

```python
print(total_pixels)
```

```python
# resize the image
crop_image = cv2.resize(crop_image,(100,100))
```

```python
plt.imshow(crop_image)
plt.axis("off")
plt.title("Lamberogini")
plt.show()
```
